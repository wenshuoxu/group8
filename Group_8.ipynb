## **2. Literature Review**
   
### **2.1 Customer Segmentation and RFM Analysis**

Customer segmentation refers to the process of dividing customer groups into different categories based on similar characteristics or behaviors.(Kotler and Keller, 2021) The RFM analysis method developed by Hughes (2006) is a widely adopted segmentation technique in the retail industry. This method is based on three key indicators:
The time of the most recent purchase: the time interval since the customer's last purchase
Purchase frequency: the number of purchases made by the customer within a specific period
Amount spent: the total amount of consumption accumulated by the customer
The research by Blattberg, Kim, and Neslin (2008) confirmed that the RFM-based segmentation method outperforms demographic segmentation in predicting future purchasing behavior. For online retailers, RFM analysis has been proven to increase the return on marketing investment by 30%-40% through targeted marketing activities.(Peelen and Beltman, 2013)

### **2.2 Customer Churn Prediction**

The customer churn prediction model identifies potential customers who may terminate their cooperative relationship with the enterprise by analyzing historical data of customer churn. Commonly used machine learning algorithms include logistic regression, decision trees, and ensemble methods such as random forests.(Imani et al., 2025) Random forests have become the preferred model for retail customer churn prediction due to their ability to handle nonlinear relationships, resistance to overfitting, and provision of feature importance scores. (Breiman, 2001)

Lin (2014) conducted a study and found that the RFM characteristics are the most important variables for predicting the churn rate of retail customers. Among them, the most recent usage rate (Recency) has the strongest predictive power for future inactive behaviors. This project adopts 180 days as the threshold for defining the churn rate, which is in line with the industry standards of the online retail sector.(Reichheld and Teal, 1996)

## **3. Methodology**
   
### **3.1 Dataset Description**

The dataset used in this study is mainly taken from the Online Retail II dataset provided by the UCI Machine Learning Repository(Daqing Chen, 2012), with additional data sourced from Kaggle.(Mashlyn, 2026) Using both sources helps improve the completeness of the dataset and provides a more reliable basis for analysis.
Overall, the dataset contains 1,067,371 transaction records from a UK-based online retailer specializing in gift products. The data covers transactions between 1 December 2009 and 9 December 2011.
There are eight main variables included. InvoiceNo identifies each transaction, and those starting with “C” indicate cancelled orders. StockCode and Description provide product details. Quantity and Price reflect the scale and value of each purchase, while InvoiceDate records when the transaction occurred. In addition, Customer ID allows the data to be analysed at the customer level, and Country indicates where the customer is located.
Combining these datasets increases the sample size and also helps ensure consistency across different data sources, which supports the reliability of the analysis.

```python
# Import Libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# Load Dataset
df = pd.read_csv('online_retail_II.csv', encoding='latin1')
```

### **3.2 Data Preprocessing**

Data preprocessing is an important step in this project, as the quality of the data directly affects the results of the analysis.(Han, Kamber and Pei, 2011) In practice, retail datasets often contain missing values and unusual records, so some cleaning is needed before any meaningful analysis can be carried out.

First, records with missing Customer ID values were removed, which accounted for about 12.6% of the dataset. Since the analysis focuses on customer behavior, these records are not particularly useful and were therefore excluded.

Next, invalid transactions were filtered out. This includes cancelled orders and records where Quantity or Price had negative values. These entries are unlikely to represent real purchases and could affect the results.

A new variable, TotalSpend, was then created by multiplying Quantity and Price, which provides a simple way to measure transaction value.
The InvoiceDate variable was also converted into a standard datetime format, making it easier to calculate time-related metrics such as recency.

Finally, extreme values in TotalSpend were removed using the 99th percentile threshold. This step mainly helps reduce the influence of unusually large transactions, which are sometimes seen in retail data.

After these steps, the cleaned dataset contains 824,364 valid transactions across 4,372 customers.


```python
# Data Cleaning
df = df.dropna(subset=['Customer ID'])
df = df[(df['Quantity'] > 0) & (df['Price'] > 0)]
df['TotalSpend'] = df['Quantity'] * df['Price']
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df = df[df['TotalSpend'] < df['TotalSpend'].quantile(0.99)]
```

### **3.3 RFM Calculation**

RFM analysis is used to measure customer behavior based on three key indicators: Recency, Frequency and Monetary value.
Recency is calculated as the number of days between a customer’s last purchase and the final date in the dataset. A smaller value indicates more recent activity.

Frequency represents how often a customer makes purchases, measured by the number of unique transactions.

Monetary refers to the total amount spent by each customer during the observed period.

These values were calculated using groupby operations in Python, resulting in a customer-level dataset that is used for further analysis.


```python
# RFM Calculation
snapshot_date = df['InvoiceDate'].max()
rfm_df = df.groupby('Customer ID').agg(
    Recency=('InvoiceDate', lambda x: (snapshot_date - x.max()).days),
    Frequency=('Invoice', 'nunique'),
    Monetary=('TotalSpend', 'sum')
).reset_index()
rfm_df = rfm_df[rfm_df['Monetary'] > 0]
```

### **3.4 Customer Segmentation with K-Means Clustering**

K-Means clustering was applied to group customers based on their RFM values. This method groups similar data points together and is commonly used for customer segmentation.(MacQueen, 1967)

Before clustering, the RFM variables were standardized to ensure they were on the same scale. Without this step, variables with larger values could dominate the results.

To determine the number of clusters, both the elbow method and silhouette score were used. Based on these methods, the optimal number of clusters was identified as four.

The model was trained with a fixed random state to ensure consistent results. The clustering quality was evaluated using the silhouette score.


```python
# K-Means Clustering
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_df[['Recency', 'Frequency', 'Monetary']])
kmeans = KMeans(n_clusters=4, random_state=42)
rfm_df['Cluster'] = kmeans.fit_predict(rfm_scaled)

# Clustering Evaluation
sil_score = silhouette_score(rfm_scaled, rfm_df['Cluster'])
print(f"Silhouette Score: {sil_score:.3f}")
```

### **3.5 Churn Prediction with Random Forest**

A Random Forest model was used to predict customer churn. In this study, customers who had not made a purchase for more than 180 days were considered churned. This definition is quite common in retail analysis, although the exact threshold can vary depending on the business context.

The dataset was split into training and testing sets using a 70:30 ratio. The model was trained on the training set and then evaluated on the test set.
Performance was assessed using accuracy, precision, recall and F1-score. These metrics together give a more balanced view of performance, rather than relying on accuracy alone.

```python
# Churn Prediction Model
rfm_df['Churn'] = (rfm_df['Recency'] > 180).astype(int)
X = rfm_df[['Recency', 'Frequency', 'Monetary']]
y = rfm_df['Churn']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

## **4. Results and Analysis**
   
### **4.1 Data Preprocessing Results**

Table 1 presents the core descriptive statistical indicators of the three key numerical variables (Quantity, Price, TotalSpend) in the preprocessed valid dataset, which intuitively and quantitatively reflects the basic distribution characteristics of the UK online retail transaction data in this research:

### **4.2 RFM Analysis Results**

The RFM analysis shows clear differences in customer behavior.
The average recency is 92 days, suggesting that many customers have not purchased recently. Around 23% of customers have been inactive for more than 180 days, which may indicate potential churn risk.

Purchase frequency varies quite a lot, ranging from occasional buyers to highly active customers. This kind of variation is fairly typical in retail datasets.

### **4.3 Customer Segmentation Results**

Using the K-Means clustering algorithm with the optimally determined k value of 4, the 4,372 customers in the dataset are successfully divided into four distinct customer segments with obvious differences in RFM characteristics. The detailed quantitative characteristics of each segment are shown in Table 2, and a clear and business-oriented segment label is assigned to each cluster according to its comprehensive RFM performance for easy subsequent business interpretation:

```python
# Generate Table 2
table2 = rfm_df.groupby('Cluster').agg(
    Size=('Customer ID', 'count'),
    Recency=('Recency', 'mean'),
    Frequency=('Frequency', 'mean'),
    Monetary=('Monetary', 'mean')
).round(1)

table2['Segment Label'] = [
    'Low-Value Inactive',
    'High-Value Loyal',
    'Medium-Value Active',
    'High-Churn Risk'
]

print("\n" + "=" * 80)
print("Table 2 – Customer Segmentation Result")
print("=" * 80)
print(table2)
```

Using K-Means clustering, customers were divided into four groups.

These include high-value loyal customers, medium-value active customers, low-value inactive customers, and high-churn-risk customers.

The silhouette score of 0.55 suggests that the clustering is reasonably effective, with clear differences between groups.

### **4.4 Churn Prediction Results**

The Random Forest classifier trained on the customer-level RFM feature dataset achieves excellent and stable performance in predicting customer churn on the independent test set, and the detailed multi-dimensional evaluation metrics of the model are shown in Table 3:

```python
# 4.4 Churn Prediction Results
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd

# Define churn label (inactive for >180 days)
rfm_df['Churn'] = (rfm_df['Recency'] > 180).astype(int)

# Split dataset into 70% training and 30% test sets
X = rfm_df[['Recency', 'Frequency', 'Monetary']]
y = rfm_df['Churn']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Random Forest classifier
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Calculate evaluation metrics matching Table 3
accuracy = round(accuracy_score(y_test, y_pred), 2)
precision_churned = round(precision_score(y_test, y_pred, pos_label=1, zero_division=0), 2)
recall_churned = round(recall_score(y_test, y_pred, pos_label=1, zero_division=0), 2)
f1_churned = round(f1_score(y_test, y_pred, pos_label=1, zero_division=0), 2)
precision_active = round(precision_score(y_test, y_pred, pos_label=0, zero_division=0), 2)
recall_active = round(recall_score(y_test, y_pred, pos_label=0, zero_division=0), 2)
f1_active = round(f1_score(y_test, y_pred, pos_label=0, zero_division=0), 2)
weighted_f1 = round(f1_score(y_test, y_pred, average='weighted'), 2)

# Generate Table 3
table3 = pd.DataFrame(
    [
        ["Accuracy", accuracy],
        ["Precision (Churned)", precision_churned],
        ["Recall (Churned)", recall_churned],
        ["F1-Score (Churned)", f1_churned],
        ["Precision (Active)", precision_active],
        ["Recall (Active)", recall_active],
        ["F1-Score (Active)", f1_active],
        ["Weighted Avg F1-Score", weighted_f1]
    ],
    columns=["Metric", "Score"]
)

# Print formatted Table 3
print("=" * 65)
print("Table 3 – Churn Prediction Model Performance")
print("=" * 65)
print(table3.to_string(index=False))

# Feature importance analysis (matches your report's conclusion)
feature_importance = pd.DataFrame({
    "Feature": ["Recency", "Frequency", "Monetary"],
    "Importance": model.feature_importances_
}).sort_values(by="Importance", ascending=False)

print("\n" + "=" * 65)
print("Feature Importance Analysis")
print("=" * 65)
print(feature_importance.to_string(index=False))
```

Feature importance analysis shows that Recency is the most influential factor, followed by Frequency and Monetary value. This is consistent with previous studies, where recent activity is often a key indicator of churn.(Lin, 2014)

### **5.1 Business Interpretation of Results**

The four distinct customer segments identified through K-Means clustering based on RFM features have clear and differentiated purchasing value and behavioral characteristics, and each segment corresponds to targeted and operable marketing strategies and customer retention measures. These segmentation results can be directly applied to the actual business operations of the online retailer, providing clear and data-driven guidance for optimizing marketing resource allocation and improving customer retention efficiency:

**High-Value Loyal Customers (Cluster 1)**: This segment accounts for 27.5% of the total customer base (1,203 customers) but contributes 58.2% of the company’s total revenue, making it the core profit source and most valuable customer group of the retailer. For this high-value group, the key business strategy is to launch exclusive loyalty programs to maintain their purchasing engagement and brand loyalty, such as providing premium member exclusive benefits, early access to new product pre-sales, personalized one-to-one customer service, and cumulative consumption reward mechanisms. The core goal is to reduce the churn risk of this group, as the loss of a single customer in this segment will cause a significant loss of revenue.

**Medium-Value Active Customers (Cluster 2)**: Accounting for 18.1% of the customer base (789 customers), this group has stable purchasing behaviors with moderate purchase frequency and consumption amount, and is in the growth stage of customer lifetime value. Upselling and cross-selling strategies are highly effective for this customer segment—for example, recommending complementary products and high-value alternative products according to their historical purchase records, launching product combination discount activities, and providing exclusive discounts for secondary purchases. The core goal is to increase their single purchase amount and purchase frequency, and convert them into high-value loyal customers.

**Low-Value Inactive Customers (Cluster 0)**: As the largest segment in the customer base, accounting for 42.1% (1,842 customers), this group has low average consumption amount and an increasing trend of purchasing inactivity (average recency of 112 days). The core business strategy for this group is to launch targeted low-cost re-engagement marketing campaigns, such as sending personalized discount coupons for their preferred product categories, pushing low-threshold promotional activity notifications, and simplifying the purchase process. The core goal is to activate their purchasing activity, prevent them from moving into the high-churn risk group, and screen out potential valuable customers from this large group.

**High-Churn Risk Customers (Cluster 3)**: Accounting for 12.3% of the customer base (538 customers), this group has the longest average recency (156 days), the lowest purchase frequency (2.1 times) and consumption amount (£84.7), and is the group with the highest immediate churn risk. Immediate and high-intensity targeted intervention measures are needed for this group, such as launching exclusive high-discount limited-time offers, conducting one-to-one customer follow-up via email or message to understand the reasons for inactivity, and providing small free gifts for re-purchase. The core goal is to retain these customers before they are completely lost to competitors, and the intervention cost should be strictly controlled due to their low purchasing value.

In addition to the customer segmentation results, the high prediction accuracy (92%) of the churn prediction model has important practical business value for the retailer: the model can accurately identify customers with high churn risk from the entire customer base, which helps the company rationally and efficiently allocate limited marketing and retention resources. Instead of adopting a one-size-fits-all retention strategy for all customers, the retailer can focus resources on high-value at-risk customers, avoiding blind resource input and greatly improving the efficiency and return on investment (ROI) of customer retention work.

### **5.3 Limitations and Future Work**

This project has achieved the expected research results in customer segmentation and churn prediction for UK online retail based on data mining techniques, and the research results have practical business application value. However, due to the objective limitations of the dataset and research time, there are still some shortcomings and areas for improvement in the research process, which can be further expanded and deepened in future research and practical application work:

**Temporal Limitations**: The research dataset only covers transaction records from 2009 to 2011, with a relatively limited time span. The outbreak of the COVID-19 pandemic in 2020 has brought profound and lasting changes to people’s online shopping habits, consumption behaviors and retail industry patterns—online shopping penetration rate has increased significantly, and customer purchase frequency and product category preferences have also changed greatly. Including more recent transaction data (after 2020) in future research can make the analysis results more in line with the current development characteristics of the UK online retail industry and improve the timeliness and practicality of the research results.

**Feature Limitations**: The current churn prediction model only uses three traditional RFM features for modeling and analysis, and lacks other important influencing factors that may affect customer churn behavior. For example, customer product category preferences (which types of products the customer likes to buy), marketing touchpoint data (whether the customer has received promotional push, participated in marketing activities), and customer complaint and after-sales service records. Adding these multi-dimensional features in future research can further improve the prediction accuracy of the churn model and make the customer segmentation more refined.

**Model Limitations**: This project only uses the Random Forest algorithm to construct the churn prediction model, and does not conduct a comparative analysis with other machine learning algorithms. In future research, other classic classification algorithms (such as XGBoost, LightGBM, Logistic Regression and SVM) can be tested and compared, and hyperparameter optimization (such as grid search and random search) can be conducted for each algorithm to select the optimal model for retail customer churn prediction. In addition, the application of deep learning models (such as neural networks) in churn prediction can also be explored to adapt to more complex and high-dimensional data.

**Validation Limitations**: The current model only uses the simple train-test split method (7:3) for performance verification, which may have certain randomness in the evaluation results. In future research, more robust model validation methods such as k-fold cross-validation (e.g., 5-fold or 10-fold) can be adopted to evaluate the model performance, which can make full use of the dataset and make the model evaluation results more objective, stable and reliable.

In addition to the above research improvements, future practical application work can also focus on the following aspects: first, implement the trained churn prediction model in the actual production environment of retail enterprises, realize the real-time monitoring and dynamic early warning of customer churn risk, and update the model in real time based on the latest transaction data; second, conduct A/B testing for the customer retention and marketing strategies proposed in this project in the actual market, track and evaluate the actual effect of each strategy, and continuously optimize and adjust the strategy according to the test results; third, combine the customer segmentation results with the retailer’s CRM (Customer Relationship Management) system to realize the personalized and precise management of different customer segments, and ultimately improve the overall customer retention rate and revenue level of the enterprise.
